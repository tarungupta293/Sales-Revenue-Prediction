```python
# create a function as merge_datasets which will merge the datasets with other.
def merge_datasets(df1, df2, onColumn):
    df_datasets_merge = pd.merge(df1, df2, on=onColumn, how="inner")
    return df_datasets_merge
```

```python
df_store.head(2)
```

```python
# filter the datasets into item categorical datasets.
filtered_df_hobbies = df_store[df_store['item_id'].str.contains('hobbies', case=False)]
filtered_df_foods = df_store[df_store['item_id'].str.contains('FOODS', case=False)]
filtered_df_household = df_store[df_store['item_id'].str.contains('household', case=False)]
```

```python
# merge the filtered datasets with the calendar dataset.
filtered_df_hobbies = merge_datasets(filtered_df_hobbies, df_calendar, ["wm_yr_wk"])    
filtered_df_foods = merge_datasets(filtered_df_foods, df_calendar, ["wm_yr_wk"])    
filtered_df_household = merge_datasets(filtered_df_household, df_calendar, ["wm_yr_wk"])
```

```python
# merge the filtered datasets with the calendar events dataset.
filtered_df_hobbies = merge_datasets(filtered_df_hobbies, df_calendar_events, ["date"])
filtered_df_foods = merge_datasets(filtered_df_foods, df_calendar_events, ["date"])
filtered_df_household = merge_datasets(filtered_df_household, df_calendar_events, ["date"])
```

```python
# filter the records having value greater than d_1541 as these records are not present in the training dataset.
filtered_df_foods = filtered_df_foods[filtered_df_foods['d'] <= 'd_1541']
```

```python
# merge the filtered dataset with the training dataset.
filtered_df_foods = merge_datasets(filtered_df_foods, df_train, ["item_id", "store_id"])
```

```python
filtered_df_household.head(2)
```

```python
# filter the 'd' column value and remove d_ from the value.
filtered_df_foods['d'] = filtered_df_foods['d'].str.extract(r'd_(\d+)')
```

```python
# transform the column d data type to int data type.
filtered_df_foods['d'] = filtered_df_foods['d'].astype(int)
```

```python
# create the new column in the dataset named as sale.
filtered_df_foods['sale'] = None
```

```python
#filtered_df_household = filtered_df_household[filtered_df_household['d'].apply(lambda x: x) <= len(filtered_df_household.columns)]
```

```python
filtered_df_foods.shape
```

```python
# insert the values in the sale column by fetching from the d column and their respective days column.
for index, row in filtered_df_foods.iterrows():
    filtered_df_foods.loc[index, 'sale'] = row[row['d'] + 12]
```

```python
filtered_df_foods.head(2)
```

```python
# Drop all columns containing d_. As these columns are not needed now.
filtered_df_foods_tmp = filtered_df_foods.drop(columns=filtered_df_foods.filter(like='d_').columns)
```

```python
filtered_df_foods_tmp.head(2)
```

```python
# create the new column as sales_revenue which will contain the sales revenue on the basis of selling price and the sale.
filtered_df_foods_tmp['sales_revenue'] = None
```

```python
# insert the values in the sales_revenue column
filtered_df_foods_tmp['sales_revenue'] = filtered_df_foods_tmp.apply(lambda row: row['sell_price'] * row['sale'], axis=1)
```

```python
# save the filtered dataset into internal storage.
filtered_df_foods_tmp.to_csv('D:\\Tarun\\UTS\\Subjects\\ADV ML\\output\\df_store_ca_04_foods_merged.csv', index = True)
```
